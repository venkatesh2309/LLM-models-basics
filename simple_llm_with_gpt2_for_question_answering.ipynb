```python
# hugging face - framework to work with llm like prompt eng,intregration with any opensource like openai,vector db etc. simple to build chatbot,Q and A
# Transformer - huggingface transformer lib for using pretained model, loads model,tokenizer etc
# accelerate - Automatically detects hardware (CPU/GPU) and configures training/inference accordingly
# huggingface_hub - A library to interact with Hugging Face Hub (the model-sharing platform). Why it’s used: Download models, upload your own, browse available models, etc.
```

```python
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from langchain.prompts import PromptTemplate

# Load the GPT-2 model and tokenizer
model_name = "gpt2"  # You can use "gpt2-medium", "gpt2-large", "gpt2-xl" if needed
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Wrap in HuggingFace pipeline for text generation
llm_pipeline = pipeline(
    "text-generation",  # GPT-2 works with text-generation pipeline
    model=model,
    tokenizer=tokenizer,
    max_length=200
)

# Convert to LangChain's HuggingFace pipeline
llm = HuggingFacePipeline(pipeline=llm_pipeline)

# Prompt template for answering a question
answer_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are a knowledgeable assistant. Based on the following passage, answer the question in a thoughtful and informative way.

Passage:
{context}

Question: {question}

Answer:
"""
)

# Create the LangChain answer chain
answer_chain = LLMChain(llm=llm, prompt=answer_prompt)

# Example context and question
context_text = """
India is a diverse and culturally rich country located in South Asia. It is the world's most populous democracy, with significant achievements in science, technology, education, and space exploration. India is also known for its vibrant traditions, festivals, and a rapidly growing economy.
"""

question_text = "How good is India?"

# Get the response using the chain
response = answer_chain.run(context=context_text, question=question_text)

print("💬", response)
```

```python
question_text = "Does India is a bad country to live?"

# Get the response using the chain
response = answer_chain.run(context=context_text, question=question_text)

print("💬", response)
```
